```python
import FACSPy as fp
```

```python
### first, we import the necessary metadata
metadata = fp.dt.Metadata("../../../Tutorials/spectral_dataset/metadata.csv")
metadata
```

```python
### view the metadata table
### sample_ID, file_name and staining are necessary columns
### everything else can be appended as the user wants
### for further information see Vignette "Supplements.ipynb"
metadata.to_df().head()
```

```python
### we also import the panel information
### the channel names have also been saved within the FCS files and could be read from the FCS File
### Here, we choose to overwrite it by our names
panel = fp.dt.Panel("../../../Tutorials/spectral_dataset/panel.txt")
panel
```

```python
panel.to_df().head()
```

```python
workspace = fp.dt.FlowJoWorkspace("../../../Tutorials/spectral_dataset/FD_NK.wsp")
workspace
```

```python
dataset = fp.create_dataset(input_directory = "../../../Tutorials/spectral_dataset/",
                            panel = panel,
                            metadata = metadata,
                            workspace = workspace,
                            subsample_fcs_to = None,
                            keep_raw = False)
```

```python
dataset
```

```python
fp.dt.calculate_cofactors(dataset)
```

```python
dataset
```

```python
### the calculated cofactors are saved in the dataset and can be extracted
calculated_cofactors = dataset.uns["cofactors"]

calculated_cofactors
```

```python
### if you have a table of your own cofactors you can load them:
supplied_cofactors = fp.dt.CofactorTable("../../../Tutorials/spectral_dataset/cofactors.txt")

supplied_cofactors
```

```python
fp.dt.transform(dataset,
                transform = "asinh",
                cofactor_table = calculated_cofactors,
                key_added = "transformed",
                layer = "compensated")
```

```python
dataset
```

```python
for marker in dataset.var_names:
    if not fp.is_fluo_channel(dataset, marker):
        continue
    fp.pl.transformation_plot(dataset,
                              sample_identifier = "2",
                              marker = marker)
```

```python
### the cofactor for CD45 is vastly overestimated
### in that case, we can set it manually:
dataset.uns["cofactors"].set_cofactor("CD45", 5000)
```

```python
### we re-transform the dataset
fp.dt.transform(dataset,
                transform = "asinh",
                cofactor_table = dataset.uns["cofactors"],
                key_added = "transformed",
                layer = "compensated")
```

```python
fp.pl.gating_strategy(dataset,
                      sample_ID = "2", 
                      wsp_group = "All Samples")
```

```python
fp.save_dataset(dataset,
                output_dir = "../../../Tutorials/spectral_dataset/",
                file_name = "raw_dataset",
                overwrite = True)
```
